```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

m = 1000

w = np.random.randn(1000)

x = np.random.uniform(0,1,m)

y = 1.2 + 2.3*x + 10*w

#t = np.arange(0.0, 1000.0, 1)


fig, ax = plt.subplots()
ax.plot(x, y, '.')

ax.set(xlabel='vetor x', ylabel='vetor y',
       title='grafico')
ax.grid()

fig.savefig("test.png")
plt.show()
```
